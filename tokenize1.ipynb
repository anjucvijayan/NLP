```python
text='hello friends! good morning! lets study Natural language procedssing. its very interesting'
```

```python
import nltk
from nltk import word_tokenize,sent_tokenize
```

```python
token_text=sent_tokenize(text)
token_text
```

```python
text1='india is my contry'
token_text=sent_tokenize(text1)
token_text
```

```python
word_token=word_tokenize(text)
word_token
```

```python
from nltk.corpus import stopwords
stop_word=set(stopwords.words('english'))
```

```python
text1='india is my contry'
w_token=word_tokenize(text1)
remove_stopwords=[w for w in w_token if not w in stop_word]
remove_stopwords
```
